#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

# Question 1
### Read the following example and complete the exercises 

You are at the age of 21 and curious about if drinking alcohol would cause liver disease. You contacted a clinic site and got the following information.

1. Past data tells you that 10% of patients entering the clinic have liver disease (**Prior**). 

2. Five percent of the clinic’s patients are alcoholics (**Marginal Probability**). 

3. Among the patients diagnosed with liver disease, 15% are alcoholics (**Likelihood**). 

Now you wonder what is the probability of a patient from the clinic have liver disease given that the patient is alcoholic (**Posterior**)

### 1.1 

#### If A represents the event of being an alcoholic patient in the clinic and L represents the event of being a patient with liver disease in the clinic, write the bayesian representation (P(A|B)) in the following cells


Posterior = P(L|A)\
P(L|A) = (P(A|L)\*P(L)) / P(A)

Prior = P(L) = 0.1

Likelihood = P(A|L) = (P(L|A)\*P(A)) / P(L) = 0.15

Marginal Probability = P(A) = 0.05

### 1.2 

#### Define all the information available and calculate the posterior probability

```julia
P_L = 0.10  
P_A_given_L = 0.15 
P_A = 0.05 

P_L_given_A = (P_A_given_L * P_L) / P_A
println("Posterior probability = P(L|A) = ", P_L_given_A)
```

Assume 1% people have Covid. If randomly select a person and he/she tests positive using the rapid antigen test, what is the probability the person has Covid (posterior).

### 2.1 

#### The probability of which event is considered marginal probability in this case? Define it in the first cell below and calculate it in the second cell below

The event is: 

```julia
P_A = 0.01 #prior probability of covid
P_Ac = 1 - P_A #probability of no covid
P_B_given_A = 0.653 #sensitivity
P_Bc_given_Ac = 0.999 #specificity
P_B_given_Ac = 1 - (P_Bc_given_Ac) #probability of testing positive given no covid
P_B = (P_B_given_A*P_A) + (P_B_given_Ac*P_Ac) #probability of testing positive
println("Marginal probability of testing positive (P(B)) = ", P_B)
```

### 2.2 

#### Define all the information and calculate the probability the randomly selected person has Covid

```julia
P_A_given_B = (P_B_given_A * P_A) / P_B #probability that randomly selected person has covid given testing positive
println("Probability that randomly selected person has covid given testing positive = ", P_A_given_B)
```

### 3.1 

#### Randomly draw 1000 samples from a normal distribution with mean = 5 and std = 2

```julia
using Pkg
Pkg.add("Random")
Pkg.add("Distributions")
Pkg.add("Plots")
using Random
using Distributions
using Plots
Random.seed!(123)
normal_distribution = Normal(5, 2)
normal_samples = rand(normal_distribution, 1000)
```

### 3.2

#### Plot  the distribution above in a histogram

```julia
histogram(normal_samples, bins = 30, title = "1000 Samples from Normal Distribution", label = "30 bins", xlabel = "Value", ylabel = "Frequency")
```

### 3.3

#### Plot  the same distribution and change the bins to 50

```julia
histogram(normal_samples, bins = 50, title = "1000 Samples from Normal Distribution With 50 Bins", label = "50 bins", xlabel = "Value", ylabel = "Frequency")
```

### 3.4 

#### Choose a distribution other than normal distribution, Describe what it represents, and draw 1000 samples from the distribution and plot it using parameters of your choice

Which distribution do you pick: Exponential distribution

What does the distribution represent?\
Exponential distribution represents the probability of time between events occurring in a Poisson process. It is a continuous distribution that models the waiting time between random events happening at a constant average rate.

```julia
exponential_distribution = Exponential(1)
exponential_samples = rand(exponential_distribution, 1000)
histogram(exponential_samples, bins = 30, title = "1000 Samples from Exponential Distribution", label = "λ = 1", xlabel = "Value", ylabel = "Frequency")
```

### 3.5

#### Plot  the two distributions in the same plot

```julia
histogram(normal_samples, bins = 30, label = "Normal (1000 samples)", xlabel = "Value", ylabel = "Frequency")
histogram!(exponential_samples, bins = 30, color = :green, label = "Exponential (1000 samples)", title = "Comparison of Normal and Exponential Distributions (1000 samples)")
```
