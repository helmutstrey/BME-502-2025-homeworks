#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

Shown by a <a href="https://pubmed.ncbi.nlm.nih.gov/34242764/#:~:text=The%20overall%20sensitivity%20of%20the,%25%20CI%2024.4%2D65.1">reaserch study</a>, the rapid antigen test for Covid has a sensitivity of 65.3% and a specificity of 99.9%. Assume 1% people have Covid. If randomly select a person and he/she tests positive using the rapid antigen test, what is the probability the person has Covid (posterior).

### 2.1 

#### The probability of which event is considered marginal probability in this case? Define it in the first cell below and calculate it in the second cell below

The event is: 

```julia
# The marginal probability would be the probability of testing positivie for 
# covid regardless of whether or not the person has covid.

# P(Positive) = P(Positive|Covid)P(Covid) + P(Positive|No Covid)P(No Covid)
```

### 2.2 

#### Define all the information and calculate the probability the randomly selected person has Covid

```julia
# Given data:
PCovid = 0.01  # Probability of having COVID (P(C))
sensitivity_T_given_C = 0.653  # Sensitivity: Probability of testing positive given COVID (P(T+|C))
specificity_T_given_not_C = 0.999  # Specificity: Probability of testing negative given no COVID (P(T-|not C))

# Compute complementary probabilities
pNoCovid = 1 - PCovid  # Probability of not having COVID
false_positive_rate = 1 - specificity_T_given_not_C  # Probability of a false positive

# Compute marginal probability of testing positive (P(T+))
marginal_T_positive = (sensitivity_T_given_C * PCovid) + (false_positive_rate * pNoCovid)

println("Marginal Probability: ", marginal_T_positive)

# Compute posterior probability using Bayes' Theorem
posterior_C_given_T_positive = (sensitivity_T_given_C * PCovid) / marginal_T_positive

println("Posterior: ", posterior_C_given_T_positive)
```
