#### Bayes' theorem:
\begin{equation*}
P(A|B) = \frac{P(B|A)P(A)}{P(B)} \\
P(A|B): Posteior \\
P(B|A): Likelihood \\
P(A): Prior \\
P(B): Marginal Probability \\
\end{equation*}

### 2.1 

#### The probability of which event is considered marginal probability in this case? Define it in the first cell below and calculate it in the second cell below

The event is: Positive testing using the rapid antigen test.

```julia
spec = 0.999
sens = 0.653
probcov = 0.01

poscovid = spec
posnocovid = 1-poscovid

margprob = (sens*probcov) + (poscovid*posnocovid)
println("The marginal probability (P(P)) is 0.753%.")
```

### 2.2 

#### Define all the information and calculate the probability the randomly selected person has Covid
#### P(C|P) = (P(P|C)*P(C))/P(P)

```julia
prob_pos_given_cov = sens
prob_pos = margprob

prob_cov_given_pos = (prob_pos_given_cov*probcov)/prob_pos
println("The probability of a randomly-selected person having Covid (P(C|P))is 86.7%.")
```

### 3.4 

#### Choose a distribution other than normal distribution, Describe what it represents, and draw 1000 samples from the distribution and plot it using parameters of your choice

Which distribution do you pick: Bernoulli

What does the distribution represent? It is a discrete probability distribution that represents two outcome, either success or failure.

```julia
using Random, Distributions

Random.seed!(123) # Setting the seed

p = 0.7 #Probability of how many are true
d = Bernoulli(p)

x = rand(d, 1000)

histogram(x, bins=2, xticks=[0,1], xlabel="Outcome", ylabel="Frequency", title="Histogram of Bernoulli(0.7)", color=:green, bar_width=0.5)
```

### 3.5

#### Plot  the two distributions in the same plot

```julia
using Distributions, StatsPlots

# Generate 1000 samples from Bernoulli(0.7)
p = 0.7
bern_dist = Bernoulli(p)
bern_samples = rand(bern_dist, 1000)

# Generate 1000 samples from Normal(0,1)
norm_dist = Normal(0,1)
norm_samples = rand(norm_dist, 1000)

# Plot histograms
histogram(bern_samples, bins=2, xticks=[0,1], xlabel="Value", ylabel="Frequency", 
          label="Bernoulli(0.7)", color=:green, bar_width=0.5, alpha=0.6, normalize=true)

histogram!(norm_samples, bins=30, label="Normal(0,1)", color=:blue, alpha=0.5, normalize=true)

# Title
title!("Comparison of Bernoulli and Normal Distributions")
```

### 3.6

#### Draw 10000 samples from the two distributions separately and plot them in the same plot

```julia
using Distributions, StatsPlots

# Generate 10,000 samples from Bernoulli(0.7)
p = 0.7
bern_dist = Bernoulli(p)
bern_samples = rand(bern_dist, 10_000)

# Generate 10,000 samples from Normal(0,1)
norm_dist = Normal(0,1)
norm_samples = rand(norm_dist, 10_000)

# Plot histograms
histogram(bern_samples, bins=2, xticks=[0,1], xlabel="Value", ylabel="Frequency", 
          label="Bernoulli(0.7)", color=:green, bar_width=0.5, alpha=0.6, normalize=true)

histogram!(norm_samples, bins=30, label="Normal(0,1)", color=:blue, alpha=0.5, normalize=true)

# Title
title!("Comparison of Bernoulli(0.7) and Normal(0,1) Distributions (10,000 Samples)")
```
